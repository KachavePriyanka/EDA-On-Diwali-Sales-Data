##Data Preparation and Cleaning
1. Load the file using Pandas
2. Look at some information about the data and the columns
3. Fix any missing or incorrect values.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding='unicode_escape')
df.head(10)
```

```python
df.columns
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True) #axis= 1 for rows and 0 for columns ....inplace implement karta hai
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.shape #12 rows are drop
```

```python
df.isnull().sum()
```

```python
#change data type ---> Amount column containing float values so change data type into int
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
#rename column name ---> not saved using inplace just for practice
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df.describe()
```

```python
#use describe for specific columns
df[['Age','Amount','Orders']].describe()
```

##Exploratory Data Analysis

```python
df.columns
```

```python
sns.countplot(x='Gender',data=df) #here values are not displayed
```

```python
a=sns.countplot(x='Gender',data=df)
for bars in a.containers:  #used for getting values
  a.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

From above barplot graph we can see that most of the buyers are females and even the purchasing power of females are greater than man


# Age

```python
df.columns
```

```python
sns.countplot(data=df,x='Age Group',hue='Gender')
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_age,x='Age Group',y='Amount')
```

From above graphs we can see that most of the buyers are of age between 26-35 Feamle

**State**

```python
#toatal number of orders from top 10 states
sales_age=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10) #head(10) will give top 10 state
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data=sales_age,x='State',y='Orders')
```

```python
#toatal amount/sales from top 10 states
sales_age=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10) #head(10) will give top 10 state
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data=sales_age,x='State',y='Amount')
#in place of kerala haryana is here because the cost/amount of purchasing is larger than kerala
```

##From above graphs we can see that most of the orders and total amount/sales are from uttar pradesh, Maharashtra, and karnataka respectively



**Maritial Status**

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(5,5)}) # if u change these figsize then it implement previous figsize
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender') #if u dont't mention hue output of graph looks different
```

**From above graph we can see taht most of the buyers are married women and they have high purchasing power**

# **Occupation**

```python
ax=sns.countplot(data=df,x='Occupation')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_occ,x='Occupation',y='Amount')
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation Sector

# Product Category

```python
ax=sns.countplot(data=df,x='Product_Category')
sns.set(rc={'figure.figsize':(30,10)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_occ=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_occ,x='Product_Category',y='Amount')
```

**From above graph we can see that most of the sold products are from food, clothing , electronics and footwear category**

```python
sales_state=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_state,x='Product_ID',y='Orders')
```

```python
#top 10 most sold products (same thing as above)
fig,ax1=plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
